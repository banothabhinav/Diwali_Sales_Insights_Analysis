```python
# import python libraries

import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
# loading the dataset(csv file) 
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
# # accessing total no of rows and columns 
df.shape
```

```python
# accessing first ten rows from df 
df.head(10)
```

```python
# accessing total information about df
df.info()
```

```python
# # accessing total description about df
df.describe()
```

```python
#drop unrelated/blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
# change data type as integers
df['Amount'] = df['Amount'].astype('int')
```

```python
# checking the data type of all columns
df.dtypes
```

```python
# accessing all column names
df.columns
```

```python
#rename columns
a = {
    'Relationship__Status':'Marital_Status',
    'Region':'Zone'
}
df1 = df.rename(columns=a)
df1.head()
```

```python
# use describe() for specific columns
df1[['Age', 'Orders', 'Amount']].describe()
```

# Exploratory Data Analysis

### Age

```python
# plotting a bar chart for Gender
ax = sns.countplot(data = df1, x = 'Age Group', hue = 'Gender')
```

```python
## plotting a bar chart for Gender and it's count
ax = sns.countplot(data = df1, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group in graphical representation
sales_age = df1.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

##### The plot indicate that most of the buyers are of age group between 26-35 yrs and females

### Gender

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df1)

sns.set(rc={'figure.figsize':(5,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount

sns.set(rc={'figure.figsize':(5,4)})
sales_gen = df1.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

##### The plots indicate that the majority of buyers are females, and furthermore, the purchasing power of females more that of males

### State

```python
# total number of orders from top 10 states

sales_state = df1.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# total amount/sales from top 10 states

sales_state = df1.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

##### The plot indicate that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

### Marital Status

```python
# plotting a bar chart for Marital Status and it's count
ax = sns.countplot(data = df1, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(5,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bra chart for marital_Status vs amount spent by the gender
sales_state = df1.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

##### The plot indicate that most of the buyers are married (women) and they have high purchasing power

### Occupation

```python
# plotting a bar for the count as per occupaation
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df1, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df1.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

##### The plot indicate that most of the buyers are working in IT, Healthcare and Aviation sector

### Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df1, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df1.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

##### The plot indicate that most of the sold products are from Food, Clothing and Electronics category

```python
sales_state = df1.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(10,4))
df1.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

## Conclusion:

In the demographic analysis, it is observed that women in the age group of 26 to 35 years, who are married and residing in the states of Uttar Pradesh, Maharashtra, and Karnataka, exhibit a higher likelihood of engaging in product purchases. Furthermore, this specific cohort is identified as being employed in key industries such as Information Technology, Healthcare, and Aviation. The data suggests a correlation between their professional backgrounds and their preferences for products falling within the Food, Clothing, and Electronics categories.
